```python
import numpy as np
from numpy import random
from numpy import mean
import matplotlib.pyplot as plt
import sympy as sp
N = 1000
u = random.rand(N)
#x = (-ln(u))^(1/2)#
sample = np.sqrt(-np.log(u))
# β ερώτημα#
m1=mean(sample**1)
print(f" H ροπή 1ης τάξης είναι : {m1}")
m2=mean(sample**2)
print(f" H ροπή 2ας τάξης είναι : {m2}")
m3=mean(sample**3 )
print(f" H ροπή 3ης τάξης είναι : {m3}")
m4=mean(sample**4 )
print(f" H ροπή 4ης τάξης είναι : {m4}")
# γ εώτημα Προσέγγιση της χαρακτηριστικής συνάρτησης:#
#Η χαρακτηριστική συνάρτηση μπορεί να προσεγγιστεί μέσω της σειράς Taylor:οπότε έχουμε#
t = sp.Symbol('t')  # Συμβολικό t#
def symbolic_F(t, m1, m2, m3, m4):
    F = (1 + 1j * t * m1 +
         (1j * t) ** 2 / 2 * m2 +
         (1j * t) ** 3 / 6 * m3 +
         (1j * t) ** 4 / 24 * m4)
    return sp.simplify(F)
F_symbolic = symbolic_F(t, m1, m2, m3, m4)
#δ ερώτημα#
x = np.linspace(0, np.max(sample), 2000)
# Θεωρητική πυκνότητα πιθανότητας: f(x)= 2x * exp(-x^2)
puknothta_pithanothtas = 2 * x * np.exp(-x**2)
plt.figure(figsize=(10, 5))
plt.hist(sample, bins=50,color='skyblue', density=True, alpha=0.8, label='Ιστογράμμα τυχαίου δείγματος')
plt.plot(x, puknothta_pithanothtas, 'r-', lw=2, label='Θεωρητική πυκνότητα')
plt.xlabel('x', fontsize=14)
plt.ylabel('Πυκνότητα πιθανότητας',fontsize=18, fontweight='bold')
plt.legend()
plt.grid('on')
plt.show()
```
